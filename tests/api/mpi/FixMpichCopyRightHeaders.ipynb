```python
import os
MPICH_DIR = "/Users/canlewi/software/mpich/test/mpi/"

def find_mpich_version(target_file):
    
    for subdir, dir, files in os.walk(MPICH_DIR):
        for file in files:
            if file == target_file and file != "testlist" and file != "Makefile.am" and file != "Makefile.mtest":
                f = open(subdir + '/' + file,'r')
                header_finished = False;
                correct_header = ""
                while(not header_finished):
                    line = f.readline()
                    if(line == '/* -*- Mode: C; c-basic-offset:4 ; indent-tabs-mode:nil ; -*- */\n'):
                        continue
                    elif "*/" in line:
                        correct_header += line
                        header_finished = True
                    else:
                        correct_header += line
                        
                return correct_header


for subdir, dir, files in os.walk("."):
    for file in files:
        if(file.endswith(".cc")):
            correct_header_text = find_mpich_version(os.path.splitext(file)[0]+'.c')
            f = open(subdir + '/' + file, 'r')
            my_file_str = f.read()
            f.close()
            if(wrong_header_text not in my_file_str):
                dummy=True
                #print ("File ", file, " from ", subdir, " did not match old header text.")
            elif correct_header_text != None:
                # Write a back up of the file
                fb = open(subdir + '/' + file + ".back", 'w')
                fb.write(my_file_str)
                fb.close()
                
                # Replace the header
                new_str = my_file_str.replace(wrong_header_text, correct_header_text)
                f = open(subdir + '/' + file, 'w')
                f.write(new_str)
                f.close()
            else:
                print("File: ", file, " in ", subdir, " did not have match in MPICH.")
        
    
```
